NumPy (Array Operations)

1.Create a NumPy array of shape (5, 5) with random integers between 1 and 50.
Find the row with the maximum sum.
Replace all elements greater than the mean with 0.


2.Create a NumPy array with numbers from 1 to 20.
Select every 3rd element starting from the 2nd element and store them in a new array.

```python
import numpy as np

# Creating 5x5 array with random integers between 1 and 50
arr = np.random.randint(1, 51, size=(5, 5))
print("Original Array:\n", arr)

# To find the row with the maximum sum
row_sums = arr.sum(axis=1)
max_sum_row_index = np.argmax(row_sums)

print("\nRow sums:", row_sums)
print("Row with maximum sum (index):", max_sum_row_index)
print("Row with maximum sum:", arr[max_sum_row_index])

# Replace all elements greater than the mean with 0
mean_value = arr.mean()
arr[arr > mean_value] = 0

print("\nMean value:", mean_value)
print("Modified Array:\n", arr)
```

```python
# Create array from 1 to 20
arr = np.arange(1, 21)
print("Original array:", arr)

# Select every 3rd element starting from the 2nd element
new_arr = arr[1::3]
print("Selected elements:", new_arr)
```

PANDAS(DATA HANDLING)

3. Load the dataset Iris data
Display the first 10 rows and basic statistics.
Filter all rows where sepal_length > median and species is "versicolor".
Group the data by species and compute the mean petal_length.

```python
#Import PANDAS & SEABORN
import pandas as pd
import seaborn as sns

#Loading data
data = sns.load_dataset("iris")

#Printing first 10 rows and statistics
data.head(10) #first 10
data.describe() #basic statistics
```

```python
# Filter rows:
# sepal_length > median AND species == 'versicolor'
# -------------------------------
median_sepal_length = data['sepal_length'].median()

filtered_df = data[
    (data['sepal_length'] > median_sepal_length) &
    (data['species'] == 'versicolor')
]

print("Filtered rows:")
print(filtered_df)

# Group by species and compute mean petal length

mean_petal_length = data.groupby('species')['petal_length'].mean()

print("Mean petal length by species:")
print(mean_petal_length)
```

4. Add a new column named petal_ratio = petal_length / petal_width.
Find the top 5 rows with the highest petal_ratio.

```python
# Adding a new column: petal_ratio = petal_length / petal_width
data['petal_ratio'] = data['petal_length'] / data['petal_width']

# Sorting with highest petal_ratio
top_5_petal_ratio = data.sort_values(by='petal_ratio', ascending=False)

#Printing Top 5
print("Top 5 rows with highest petal_ratio:")
print(top_5_petal_ratio.head(5))
```

DATA VISUALIZATION

5. Using Matplotlib, create:
A histogram of sepal_length with 15 bins.
A scatter plot of sepal_length vs petal_length, color-coded by species.

```python
import matplotlib.pyplot as plt

#HISTOGRAM CREATION
plt.hist(data['sepal_length'], bins=15, color='r')
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.show()
```

```python
#SCATTER PLOT CREATION
species_unique = data['species'].unique()
colors = ['red', 'green', 'blue']

for sp, color in zip(species_unique, colors):
    subset = data[data['species'] == sp]
    plt.scatter(subset['sepal_length'], subset['petal_length'], label=sp, color=color)

plt.title('Sepal Length vs Petal Length by Species')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.legend()
plt.show()
```

6. Using Seaborn, plot:
A pairplot of the entire dataset.
A boxplot showing petal_width distribution for each species.

```python
#PAIR PLOT
sns.pairplot(data, hue='species', corner=True)
plt.suptitle("Pairplot of Iris Dataset", y=1.02)  # TITLE CREATION
plt.show()
```

```python
#BOXPLOT CREATION
sns.boxplot(x='species', y='petal_width', hue='species',data=data, dodge=False)
plt.title("Petal Width Distribution by Species")
plt.xlabel("Species")
plt.ylabel("Petal Width")
plt.show()
```
