```python
import numpy as np
data["Year"] = data["Year"]**2
data["Kilometer"] = np.log1p(data["Kilometer"])
data["Height"] = np.cbrt(data["Height"])
data["Seating Capacity"] = np.cbrt(data["Seating Capacity"])
data["Fuel Tank Capacity"] = np.cbrt(data["Fuel Tank Capacity"])
data["Price"] = np.log1p(data["Price"])
```

```python
data.select_dtypes("number").skew()
```

```python
data["Kilometer"] = np.expm1(data["Kilometer"]) #kilometer highly skewed so reversing log1p
```

```python
data.select_dtypes("number").skew() #checking skewness again
```

```python
# Cap outliers for kilometer
Q1 = data["Kilometer"].quantile(0.25)
Q3 = data["Kilometer"].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR
data["Kilometer"] = data["Kilometer"].clip(lower,upper)
```

```python
numcol = ['Year', 'Kilometer', 'Owner', 'Engine', 'Max Power', 'Max Torque',
       'Length', 'Width', 'Height', 'Seating Capacity', 'Fuel Tank Capacity']
```

```python
from sklearn.preprocessing import StandardScaler #Scaling num values columns
std = StandardScaler()
data[numcol] = std.fit_transform(data[numcol])
```

```python
X_final #scaling completed, FEATURES
```

```python
y= data["Price"] #TARGET
y
```

```python
from sklearn.model_selection import train_test_split #Regression methods
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.metrics import mean_absolute_error, root_mean_squared_error, r2_score,mean_squared_error
import numpy as np
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_final, y, test_size=0.2, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
#Linear regression
lin_model = LinearRegression()
lin_model.fit(X_train,y_train)
y_pred_lin = lin_model.predict(X_test)
print("Linear Regression Metrics:")
print("R2:", r2_score(y_test, y_pred_lin))
print("MAE:", mean_absolute_error(y_test, y_pred_lin))
print("MSE:", mean_squared_error(y_test, y_pred_lin))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred_lin)))
```

```python
#Ridge regression #R2 vlaue more here, so ridge is best option in Regression
ridge_model = Ridge(alpha=1.0)
ridge_model.fit(X_train, y_train)
y_pred_ridge = ridge_model.predict(X_test)

print("\nRidge Regression Metrics:")
print("R2:", r2_score(y_test, y_pred_ridge))
print("MAE:", mean_absolute_error(y_test, y_pred_ridge))
print("MSE:", mean_squared_error(y_test, y_pred_ridge))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred_ridge)))
```

```python
#lasso  #r2 score lesser than ridge, ridge is the best
lasso_model = Lasso(alpha=0.01)
lasso_model.fit(X_train, y_train)
y_pred_lasso = lasso_model.predict(X_test)

print("\nLasso Regression Metrics:")
print("R2:", r2_score(y_test, y_pred_lasso))
print("MAE:", mean_absolute_error(y_test, y_pred_lasso))
print("MSE:", mean_squared_error(y_test, y_pred_lasso))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred_lasso)))
```
