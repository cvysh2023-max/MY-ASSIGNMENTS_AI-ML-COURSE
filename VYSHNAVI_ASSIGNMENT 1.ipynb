Simple ATM Calculator

```python
balance = float(input("Enter your account balance: "))
withdrawal = float(input("Enter the amount you want to withraw: "))
if balance > withdrawal:
  balance = balance - withdrawal
  print("Your new balance after withdrawal is", balance)
else:
  print("Insufficient balance to withdraw")
```

Bus Seat Reservation System

```python


#Take initial bookings
n = int(input("How many passengers booked initially? "))
booked = []

for i in range(n):
    name = input(f"Enter name {i+1}: ")
    booked.append(name)

# Cancellation
cancel = input("Enter the name to cancel: ")
if cancel in booked:
    booked.remove(cancel)
else:
    print("Name not found in the list.")

# New booking
new = input("Enter the new passenger name: ")
position = int(input("Enter the position to insert (0 for first): "))
booked.insert(position, new)

# Show updated list
print("Updated booking list:", booked)
```

Dictionary – Student Marks Summary

```python

# Take sentence input
sentence = input("Enter the input sentence: ")

# Split sentence into words
words = sentence.split()

# Create an empty dictionary for word frequency
freq = {}

# Count each word
for word in words:
    if word in freq:
        freq[word] += 1
    else:
        freq[word] = 1

# Display the frequency
for word, count in freq.items():
    print(f"{word}: {count}")
```

Functions – Check Prime

```python

# Function to check if a number is prime
def is_prime(num):
    if num <= 1:  # Numbers less than or equal to 1 are not prime
        return False
    for i in range(2, num):  # Check divisibility from 2 to num-1
        if num % i == 0:     # If divisible, it's not prime
            return False
    return True  # If no divisors found, it's prime

# Take user input
number = int(input("Enter a number: "))

# Call the function and display result
if is_prime(number):
    print(f"{number} is a prime number.")
else:
    print(f"{number} is not a prime number.")

```

Vowel Counter

```python
def count_vowels(text):
  count = 0
  vowels = "aeiouAEIOU"
  for char in text:
    if char in vowels:
      count = count + 1
  return count


sentence = str(input("Enter the input sentence: "))

print("Number of vowels is:", count_vowels(sentence))
```
